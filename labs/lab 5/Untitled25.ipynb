```python
import numpy as np

sensor_data = np.array([
    [12, 15, 14, 10, 13],
    [20, 22, 19, 18, 21],
    [30, 28, 27, 29, 31],
    [25, 24, 26, 23, 22]
])

avg_per_machine = np.mean(sensor_data, axis=1)
print("Average per machine:", avg_per_machine)
max_per_column = np.max(sensor_data, axis=0)
print("Max per column:", max_per_column)

normalized_data = sensor_data / max_per_column
print("Normalized data:\n", normalized_data)

machines_above_20 = np.where(avg_per_machine > 20)[0]
print("Machines with avg > 20:", machines_above_20)

reshaped_data = sensor_data.reshape(2, 10)
print("Reshaped data:\n", reshaped_data)
```

```python
import numpy as np

A = np.random.randint(1, 51, (3, 3))
B = np.random.randint(1, 51, (3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

mat_mult = A @ B
print("\nMatrix Multiplication:\n", mat_mult)

elem_mult = A * B
print("\nElement-wise Multiplication:\n", elem_mult)

greater_100 = mat_mult[mat_mult > 100]
print("\nElements > 100:\n", greater_100)

A[A < 10] = 0
B[B < 10] = 0

print("\nMatrix A after replace:\n", A)
print("Matrix B after replace:\n", B)

stacked = np.vstack((A, B))
print("\nStacked Matrix:\n", stacked)
```

```python
import pandas as pd

sales_data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Customer": ["Ali", "Sara", "Ahmed", "Ali", "Zain", "Sara"],
    "City": ["Karachi", "Lahore", "Karachi", "Islamabad", "Lahore", "Karachi"],
    "Category": ["Electronics", "Clothing", "Electronics", "Groceries", "Clothing", "Electronics"],
    "Amount": [12000, 3500, None, 2400, 5200, 15000]
}

df = pd.DataFrame(sales_data)

print("First 3 rows:\n", df.head(3))

print("\nMissing values:\n", df.isnull())

median_value = df["Amount"].median()
df["Amount"].fillna(median_value, inplace=True)

df["Amount"] = df["Amount"].astype(int)

print("\nCleaned Data:\n", df)

df.to_csv("cleaned_sales_data.csv", index=False)

print("\nFile saved as cleaned_sales_data.csv")
```

```python
import pandas as pd

sales_data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Customer": ["Ali", "Sara", "Ahmed", "Ali", "Zain", "Sara"],
    "City": ["Karachi", "Lahore", "Karachi", "Islamabad", "Lahore", "Karachi"],
    "Category": ["Electronics", "Clothing", "Electronics", "Groceries", "Clothing", "Electronics"],
    "Amount": [12000, 3500, 8000, 2400, 5200, 15000]
}

df = pd.DataFrame(sales_data)

df["HighValue"] = df["Amount"] > 10000
print("With HighValue column:\n", df)

karachi_orders = df[df["City"] == "Karachi"]
print("\nKarachi Orders:\n", karachi_orders)

electronics_high = df[(df["Category"] == "Electronics") & (df["Amount"] > 10000)]
print("\nElectronics with Amount > 10000:\n", electronics_high)

sorted_df = df.sort_values(by="Amount", ascending=False)
print("\nSorted Data:\n", sorted_df)

top_customers = df.groupby("Customer")["Amount"].sum().sort_values(ascending=False).head(2)
print("\nTop 2 Customers:\n", top_customers)
```

```python
import pandas as pd

sales_data = {
    "OrderID": [101, 102, 103, 104, 105, 106],
    "Customer": ["Ali", "Sara", "Ahmed", "Ali", "Zain", "Sara"],
    "City": ["Karachi", "Lahore", "Karachi", "Islamabad", "Lahore", "Karachi"],
    "Category": ["Electronics", "Clothing", "Electronics", "Groceries", "Clothing", "Electronics"],
    "Amount": [12000, 3500, 8000, 2400, 5200, 15000]
}

df = pd.DataFrame(sales_data)

revenue_city = df.groupby("City")["Amount"].sum()
print("Revenue per City:\n", revenue_city)

avg_category = df.groupby("Category")["Amount"].mean()
print("\nAverage per Category:\n", avg_category)

unique_customers = df.groupby("City")["Customer"].nunique()
print("\nUnique Customers per City:\n", unique_customers)

pivot = pd.pivot_table(df, values="Amount", index="City", columns="Category", aggfunc="sum")
print("\nPivot Table:\n", pivot)

revenue_city_df = revenue_city.reset_index()
print("\nRevenue with reset index:\n", revenue_city_df)
```
