```python
import requests
from bs4 import BeautifulSoup
url = "http://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
quotes_data = []

quotes = soup.find_all("div", class_="quote")

for q in quotes:
    text = q.find("span", class_="text").text
    author = q.find("small", class_="author").text
    data = {
        "quote": text,
        "author": author
    }
    quotes_data.append(data)
for item in quotes_data:
    print(item)
```

```python
import requests
from bs4 import BeautifulSoup

all_quotes = []
for page in range(1, 4):
    url = f"http://quotes.toscrape.com/page/{page}/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    quotes = soup.find_all("div", class_="quote")

    for q in quotes:
        text = q.find("span", class_="text").text
        author = q.find("small", class_="author").text

        data = {
            "quote": text,
            "author": author
        }
        all_quotes.append(data)
print("Total quotes collected:", len(all_quotes))
print("\nFirst 5 quotes:\n")
for i in range(5):
    print(all_quotes[i])
```

```python
import sqlite3
conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS quotes (
    id INTEGER PRIMARY KEY,
    quote TEXT,
    author TEXT
)
""")
conn.commit()
print("Table 'quotes' is ready!")
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()
data_to_insert = []
for item in all_quotes:
    data_to_insert.append((item["quote"], item["author"]))
cursor.executemany(
    "INSERT INTO quotes (quote, author) VALUES (?, ?)",
    data_to_insert
)
conn.commit()
print("All quotes inserted successfully!")
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()
cursor.execute("SELECT quote, author FROM quotes")
rows = cursor.fetchall()
print("Total quotes in database:", len(rows))
print("\nFirst 10 quotes:\n")

for i in range(min(10, len(rows))):
    print(rows[i])
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()
author_name = "Albert Einstein"

cursor.execute(
    "SELECT quote FROM quotes WHERE author = ?",
    (author_name,)
)

results = cursor.fetchall()

print(f"\nQuotes by {author_name}:\n")
for q in results:
    print(q[0])
cursor.execute("""
SELECT author, COUNT(*)
FROM quotes
GROUP BY author
""")

counts = cursor.fetchall()

print("\nQuotes count for each author:\n")
for item in counts:
    print(item)
cursor.execute("""
SELECT author, COUNT(*) as total
FROM quotes
GROUP BY author
ORDER BY total DESC
""")

sorted_authors = cursor.fetchall()

print("\nAuthors sorted by number of quotes:\n")
for item in sorted_authors:
    print(item)
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()
author_name = "Albert Einstein"
cursor.execute(
    "UPDATE quotes SET author = UPPER(author) WHERE author = ?",
    (author_name,)
)
conn.commit()

print(f"Author '{author_name}' updated to uppercase!")
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect("quotes.db")
cursor = conn.cursor()
cursor.execute(
    "DELETE FROM quotes WHERE LENGTH(quote) < ?",
    (30,)
)
conn.commit()

print("Short quotes deleted successfully!")
conn.close()
```
