```python
import pandas as pd
import numpy as np
data = {
    "id": [1, 2, 3, 4, 5, 6],
    "name": ["Ali", "Sara", "Ahmed", "Fatima", "Usman", None],
    "age": [20, 22, None, 23, 24, 22],
    "experience": [1, 2, None, 3, 2, None],
    "expected_salary": [50000, 60000, "?", 55000, 65000, 58000],
    "education": ["BS", "MS", "BS", None, "MS", None]
}
df = pd.DataFrame(data)
df["age"].fillna(df["age"].median(), inplace=True)
df["experience"].fillna(0, inplace=True)
df["expected_salary"] = df["expected_salary"].replace("?", np.nan)
df["expected_salary"] = df["expected_salary"].astype(float)

df["expected_salary"].fillna(df["expected_salary"].median(), inplace=True)
df["education"].fillna("Unknown", inplace=True)
df["name"].fillna("Unknown", inplace=True)
print("Missing values after cleaning:\n")
print(df.isnull().sum())

print("\nCleaned Data:\n")
print(df)
```

```python
import pandas as pd

data = {
    "age": [20, 22, 21, 23, 24, 22],
    "experience": [0, 2, 1, 3, 6, 4],
    "expected_salary": [50000, 60000, 55000, 52000, 90000, 65000],
    "experience_level": ["junior", "mid", "junior", "mid", "senior", "mid"]
}
df = pd.DataFrame(data)
cols_to_normalize = ["age", "experience", "expected_salary"]
for col in cols_to_normalize:
    min_val = df[col].min()
    max_val = df[col].max()

    df[col] = (df[col] - min_val) / (max_val - min_val)
print(df)
```

```python
import pandas as pd
import numpy as np
def preprocess_applicants(df):
    df = df.copy()
    if "age" in df.columns:
        df["age"] = df["age"].fillna(df["age"].median())
    if "experience" in df.columns:
        df["experience"] = df["experience"].fillna(0)
    if "expected_salary" in df.columns:
        df["expected_salary"] = df["expected_salary"].replace("?", np.nan)
        df["expected_salary"] = df["expected_salary"].astype(float)
        df["expected_salary"] = df["expected_salary"].fillna(df["expected_salary"].median())

    if "education" in df.columns:
        df["education"] = df["education"].fillna("unknown")
    if "name" in df.columns:
        df["name"] = df["name"].fillna("unknown")
    if "city" in df.columns:
        df["city"] = df["city"].str.lower()
        mapping = {
            "isl": "islamabad",
            "isb": "islamabad"
        }
        df["city"] = df["city"].replace(mapping)
    if "experience" in df.columns:
        bins = [-1, 1, 5, float("inf")]
        labels = ["junior", "mid", "senior"]

        df["experience_level"] = pd.cut(df["experience"], bins=bins, labels=labels)
    cols_to_norm = ["age", "experience", "expected_salary"]

    for col in cols_to_norm:
        if col in df.columns:
            min_val = df[col].min()
            max_val = df[col].max()

            if max_val != min_val:
                df[col] = (df[col] - min_val) / (max_val - min_val)
    return df
data = {
    "name": ["Ali", "Sara", None],
    "age": [20, None, 22],
    "experience": [1, None, 6],
    "expected_salary": [50000, "?", 80000],
    "education": ["BS", None, "MS"],
    "city": ["isl", "Lahore", "karachi"]
}

df_raw = pd.DataFrame(data)

df_clean = preprocess_applicants(df_raw)

print(df_clean)
```
