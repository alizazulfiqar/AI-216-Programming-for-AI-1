```python
prices=[1200,850,4300,2999,1500,850]

total_prices=len(prices)
print("the total_prices:",total_prices)
average_prices=sum(prices)/len(prices)
print("the average of prices:",average_prices)
above_prices=[]
for price in prices:
  if price>2000:
    above_prices.append(price)
print("the above_prices:",above_prices)

prices.sort(reverse=True)
print("sorting prices in decending:",prices)
```

```python
employees = {
    "E01": {"name": "Ali", "salary": 60000},
    "E02": {"name": "Sara", "salary": 75000},
    "E03": {"name": "Ahmed", "salary": 50000}
}
print("employees ids",employees.keys())
total_payroll=0
for emp in employees:
  total_payroll+=employees[emp]["salary"]
print("the total_payroll:",total_payroll)
high_earns=[]
for emp_id,emp_info in employees.items():
  if emp_info["salary"]>60000:
    high_earns.append(emp_info["name"])
print("the employees earning above 60000:",high_earns)
employees["E04"]={"name":"aliza","salar":80000}
print(employees)

```

```python
sales_data = {
    "electronics": [
        ("Laptop", 120000, 3),
        ("Mouse", 1500, 10)
    ],
    "clothing": [
        ("Shirt", 2500, 5),
        ("Shoes", 5000, 2)
    ]
}
category_revenue = {}

unique_products = set()

for category, products in sales_data.items():
    total = 0
    for product_name, price, quantity in products:
        total += price * quantity
        unique_products.add(product_name)
    category_revenue[category] = total
print("Revenue Report")
for category, revenue in category_revenue.items():
    print(f"Category: {category.capitalize()}")
    print(f"  Total Revenue: {revenue}")
print("Unique Products Sold:", ", ".join(unique_products))
```
