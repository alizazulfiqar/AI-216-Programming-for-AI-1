```python
rent=12000
food=25000
transport=4000
utilities=15000
budget_limit=60000
total_expense=rent+food+transport+utilities
print("total_expense:",total_expense,"pkr")
if total_expense<=budget_limit:
  print("status:within budget")
else:
  print("status:over budget")
```

```python
attendence=[72,85,90,60,40,78,88]
meet_attendence=0
not_meet_attendence=0
for percent in attendence:
  if percent>=75:
    meet_attendence+=1
  else:
    not_meet_attendence+=1
print("students meeting attendence:",meet_attendence)
print("students not meeting attendence:",not_meet_attendence )
```

```python
cgpas=[3.2,3.9,3.8,3.1,3.0,3.7]
family_income=[70000,90000,60000,50000,80000,30000]
eligibile_count=0
for i in range(len(cgpas)):
   if cgpas[i]>=3.5 and family_income[i]<=80000:
    print("student",i+1,"eligilibile for scholarship")
    eligibile_count+=1
   else:
       print("student",i+1,"is not eligibile for scholarship")

print("total eligibile student:",eligibile_count)
```

```python
readings=[25,13,75,45,55,65,34]
lower_limit=20
upper_limit=80
alert_count=0
for value in readings:
  if value < lower_limit or value > upper_limit:
    alert_count+=1
    print(value,"alert in safe range")
  else:
    print(value,"normal")

total_readings=len(readings)
alert_percentage=(alert_count/total_readings*100)
print("total alerts:",alert_count)
print("alert percentage:",alert_percentage)
```
